# melt/pivot 整理数据

## melt函数功能演示

##### 作用：将宽数据变长数据
**pd.melt(df, id_vars=None, value_vars=None, var_name=None, value_name='value')**<br/>或<br/>
**df.melt(id_vars=None, value_vars=None, var_name=None, value_name='value')**<br/>
参数：<br/>
* df：DataFrame数据<br/>
* id_vars：固定不动的列<br/>
* value_vars：需要被转换的列，如果不指定该参数，除id_vars之外的列都会进行转换

**加载数据 pew.csv 数据**

```python
pew = pd.read_csv('./data/pew.csv')
pew.head()
```

```python
pew_long = pd.melt(pew, id_vars=['religion'])
pew_long
```

```python
pew_long = pd.melt(pew, id_vars=['religion'], var_name='income', value_name='count')
pew_long.head()
```

## pivot 函数功能演示

##### 作用：将长数据变宽数据
**pd.pivot(df, index=None, columns=None, values=None)**<br/>
或<br/>
**df.pivot(index=None, columns=None, values=None)**<br/>
参数：
* df：DataFrame数据
* index：指定列标签，该列的值作为结果中的行标签
* columns：指定列标签，该列的值作为结果中的列标签
* values：指定列标签，该列的值作为结果中每个单元格的值

```python
pew_wide = pew_long.pivot(index='religion', columns='income', values='count')
pew_wide
```

```python
# 重置行标签
pew_wide.reset_index()
```

# stack/unstack 整理数据

## stack 功能演示

##### 作用：将行转列，同时列标签转成内层行标签
**df.stack()**

```python
state_fruit = pd.read_csv('./data/state_fruit.csv', index_col=0)
state_fruit
```

```python
# stack操作
state_fruit_series = state_fruit.stack()
state_fruit_series
```

```python
state_fruit_tidy = state_fruit_series.reset_index()
state_fruit_tidy
```

```python
state_fruit_tidy.columns = ['state', 'fruit', 'weight']
state_fruit_tidy
```

# wide_to_long 整理数据

##### 作用：宽数据变长数据，能够把列名起始部分相同的列放在一起进行转换
**pd.wide_to_long(df, stubnames, i, j, sep='')**<br/>
参数：
* df：DataFrame数据
* stubnames：转换列的起始部分(前缀)
* i：固定不动的列，作为转换结果的行标签
* j：转化列后缀值组成的列的列标签名
* sep：转换列的前缀和后缀的分割符

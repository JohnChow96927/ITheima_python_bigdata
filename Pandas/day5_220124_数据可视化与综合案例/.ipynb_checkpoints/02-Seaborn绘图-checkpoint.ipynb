```python
import seaborn as sns
```

# 加载数据

```python
tips = sns.load_dataset('tips')
tips
```

# 单变量绘图

## 直方图 vs 密度图

**示例：查看客户消费金额 total_bill 的分布情况**

```python
# 直方图
sns.histplot(tips, x='total_bill')
```

```python
# 密度图
sns.kdeplot(tips['total_bill'])
```

```python
# 直方图 vs 密度图
sns.histplot(tips, x='total_bill', bins=10, kde=True)  # bins: 分多少个区间
```

## 箱线图 vs 小提琴图

**示例：查看客户消费金额 total_bill 的分布情况**

```python
# 箱线图
sns.boxplot(y='total_bill', data=tips)
```

```python
# 小提琴图
sns.violinplot(y='total_bill', data=tips)
```

## 条形图 vs 计数条形图

**示例：按照 day(周几) 统计每天的消费数量**

```python
# 条形图
result = tips['day'].value_counts()
result
```

```python
sns.barplot(x=result.index, y=result.values)
```

```python
# 计数条形图
sns.countplot(x='day', data=tips)
```

# 双变量绘图

## 散点图 vs 蜂巢图

**示例：查看消费数据中消费 tip 随着消费金额 total_bill 的变化情况**

```python
# 散点图
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
# 蜂巢图
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

## 条形图 vs 折线图

**示例：按 day(周几) 统计每餐消费的平均值并绘图**

```python
# 条形图
result = tips.groupby('day').total_bill.mean()
result
```

```python
sns.barplot(x=result.index, y=result.values)
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
```

**示例：按 day(周几) 统计每餐消费的平均值并绘图**

```python
# 折线图
result = tips.groupby('day').total_bill.mean()
result
```

```python
result = tips.groupby('day').total_bill.mean()

sns.lineplot(x=result.index, y=result.values)
```

```python
sns.lineplot(x='day', y='total_bill', data=tips)
```

# 多变量绘图

## 散点图 vs 折线图

```python
# 散点图：查看不同性别的消费客户中，消费 tip 随着消费金额 total_bill 的变化情况
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
# 折线图：查看不同性别的消费客户中，消费金额 total_bill 随着 day 的变化情况
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8, 7))
axes = fig.subplots(1, 1)

sns.lineplot(x='day', y='total_bill', data=tips, hue='sex', ax=axes)
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar

# 按 day(周几) 统计每餐消费的平均值
result = tips.groupby('day').total_bill.mean()

c = (
    Bar()
    .add_xaxis(result.index.tolist()) # 需要将 result.index 转换为列表
    .add_yaxis("平均消费金额", result.values.tolist()) # 需要将 result.values 转换为列表
    .set_global_opts(
        title_opts=opts.TitleOpts(title="用餐平均消费(按周几)"),
        yaxis_opts=opts.AxisOpts(name="平均消费金额"),
        xaxis_opts=opts.AxisOpts(name="周几"),
    )
)
# 在 jupyter notebook 中绘制图形，显示需要调用 render_notebook 方法
c.render_notebook()
```
